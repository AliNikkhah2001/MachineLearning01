## Get the data

Get the data and split them into training, validation, and test sets, by running the following cell. `X` is the input images, `Y` is the ground-truth for each input, and `y` is the input's label.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
from ae_helper import get_data
from sklearn.model_selection import train_test_split
```

```python
# Load and prepare data
X, Y, y = get_data()
X_train, X_test, Y_train, Y_test, y_train, y_test = train_test_split(X, Y, y, test_size=0.2, random_state=17)
X_train, X_val, Y_train, Y_val, y_train, y_val = train_test_split(X_train, Y_train, y_train, test_size=0.1, random_state=17)

# Normalize and reshape the data correctly for PyTorch (if not already done in get_data)
# Assuming X_train, X_val, X_test are in range [0, 1] and are numpy arrays of shape (n, 28, 28)
X_train, X_val, X_test = [x.reshape(-1, 1, 28, 28) for x in (X_train, X_val, X_test)]
```

## Design your model

Please Implement your autoencoder model below. You MUST implement your model using the `torch` package. The model should input a batch of images and output the generated images for the whole batch. Your model can be consist of fully connected layers (using linear layers), but using fully-convolutional models has extra points.

```python
class ConvAutoencoder(nn.Module):
    def __init__(self):
        super(ConvAutoencoder, self).__init__()
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, 3, stride=2, padding=1),  # [16, 14, 14]
            nn.ReLU(),
            nn.Conv2d(16, 32, 3, stride=2, padding=1),  # [32, 7, 7]
            nn.ReLU(),
            nn.Conv2d(32, 64, 3, stride=2, padding=1),  # [64, 4, 4]
            nn.ReLU(),
            nn.Conv2d(64, 128, 3, stride=1),  # [128, 2, 2]
            nn.ReLU(),
            nn.Conv2d(128, 256, 2),  # [256, 1, 1]
            nn.ReLU()
        )
        # Decoder
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 2),  # Output: [128, 2, 2]
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, 3, stride=2),  # Output: [64, 5, 5]
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, 3, stride=2),  # Output: [32, 11, 11]
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, 4, stride=2),  # Output: [16, 24, 24]
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, 4, stride=2,padding=11),  # Output: [1, 28, 28]
            nn.Sigmoid()

        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
test_input = torch.randn(1, 1, 28, 28)  # Batch size of 1, 1 channel, 28x28 pixels
model = ConvAutoencoder()
output = model(test_input)
print("Output size:", output.size())  # Should be [1, 1, 28, 28]
```

## The dataset class, and the dataloaders

The following cells implement the dataset class and create the data loaders for training and validation sets. Feel free to modify the batch size for each data loader.

```python
class MnistNextDigitDataset(Dataset):
    def __init__(self, X: np.ndarray, Y: np.ndarray, y: np.ndarray):
        self.X = torch.from_numpy(X).float()
        self.Y = torch.from_numpy(Y).float()  # Y is not used in your training function, consider this
        self.y = torch.from_numpy(y).long()

    def __len__(self) -> int:
        return len(self.X)

    def __getitem__(self, i: int) -> Tuple[torch.Tensor, ...]:
        return self.X[i], self.Y[i], self.y[i]

# DataLoader creation
train_dataloader = DataLoader(MnistNextDigitDataset(X_train, Y_train, y_train), batch_size=512, shuffle=True)
val_dataloader = DataLoader(MnistNextDigitDataset(X_val, Y_val, y_val), batch_size=1024, shuffle=False)

# ConvAutoen
```

```python
def train_autoencoder(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in dataloader:
        imgs = batch[0].to(device)  # Assuming the first element is images
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = F.mse_loss(outputs, imgs)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)
```

```python
def evaluate_autoencoder(model, dataloader, device):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for batch in dataloader:
            imgs = batch[0].to(device)
            outputs = model(imgs)
            loss = F.mse_loss(outputs, imgs)
            total_loss += loss.item()
    return total_loss / len(dataloader)
```

## Instantiate the model and optimizer

Instantiate your model and optimizer in the following cell. Feel free to choose the optimizer and its configurations such as learning rate, decay, etc.

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = ConvAutoencoder().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
print(model)
```

## Train your autoencoder

Write your training/validation loop in the following cell for an arbitrary number of epochs (until convergence is detected). Make sure to use proper and readable logging containing training and validation metrics for each epoch. Select and store the state dictionary of the best epoch model according to the metrics.

```python
# Execute training loop
num_epochs = 5
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    train_loss = train_autoencoder(model, train_dataloader, optimizer, device)
    val_loss = evaluate_autoencoder(model, val_dataloader, device)
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    print(f"Epoch {epoch+1}, Train Loss: {train_loss}, Val Loss: {val_loss}")
```

```python
import matplotlib.pyplot as plt
import random

def plot_samples(model, data_loader, device, num_samples=10):
    """
    Plot samples of original and decoded images from the autoencoder.

    Args:
    model (nn.Module): The autoencoder model.
    data_loader (DataLoader): DataLoader containing the dataset to plot.
    device (torch.device): Device to perform computations on.
    num_samples (int): Number of samples to plot.
    """
    model.eval()
    fig, axs = plt.subplots(2, num_samples, figsize=(20, 4))  # Two rows: input, output

    # Get a batch of data
    for inputs, outputs, labels in data_loader:
        inputs, outputs = inputs.to(device), outputs.to(device)
        predicted = model(inputs)
        predicted = predicted.detach().cpu().numpy()
        inputs = inputs.cpu().numpy()
        outputs = outputs.cpu().numpy()

        for i in range(num_samples):
            if i >= inputs.shape[0]:  # Check if the batch has enough samples
                break
            # Random color map selection from matplotlib's available colormaps
            random_cmap = random.choice([
                'viridis', 'plasma', 'inferno', 'magma', 'cividis',
                'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'
            ])

            # Plot input image
            axs[0, i].imshow(inputs[i].squeeze(), cmap='gray')
            axs[0, i].set_title(f'Input Label: {labels[i]}')
            axs[0, i].axis('off')

            # Plot output image
            axs[1, i].imshow(predicted[i].squeeze(), cmap=random_cmap)
            axs[1, i].set_title(f'Output Label: {labels[i] + 1}')  # Assuming label increment
            axs[1, i].axis('off')

    plt.show()

# Example usage:
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
# Ensure `data_loader` is set up to provide the correct pairs of current and next images.
plot_samples(model, val_dataloader, device)
```

# Another Conv AuroEncoder Structure

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ConvAutoencoder(nn.Module):
    def __init__(self):
        super(ConvAutoencoder, self).__init__()
        ## Encoder layers ##
        # Conv layer (depth from 1 --> 16), 3x3 kernels
        self.conv1 = nn.Conv2d(1, 16, 3, padding=1)  
        # Conv layer (depth from 16 --> 4), 3x3 kernels
        self.conv2 = nn.Conv2d(16, 4, 3, padding=1)
        # Pooling layer to reduce x-y dims by two; kernel and stride of 2
        self.pool = nn.MaxPool2d(2, 2)
        
        ## Decoder layers ##
        # A kernel of 2 and a stride of 2 will increase the spatial dims by 2
        self.t_conv1 = nn.ConvTranspose2d(4, 16, 2, stride=2)
        self.t_conv2 = nn.ConvTranspose2d(16, 1, 2, stride=2)

    def forward(self, x):
        ## Encode ##
        # Add hidden layers with relu activation function
        # and maxpooling after
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        # Add second hidden layer
        x = F.relu(self.conv2(x))
        x = self.pool(x)  # Compressed representation
        
        ## Decode ##
        # Add transpose conv layers, with relu activation function
        x = F.relu(self.t_conv1(x))
        # Output layer (with sigmoid for scaling from 0 to 1)
        x = torch.sigmoid(self.t_conv2(x))
                
        return x

# Initialize the neural network
model = ConvAutoencoder()
print(model)

```

```python
# Execute training loop
num_epochs = 5
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    train_loss = train_autoencoder(model, train_dataloader, optimizer, device)
    val_loss = evaluate_autoencoder(model, val_dataloader, device)
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    print(f"Epoch {epoch+1}, Train Loss: {train_loss}, Val Loss: {val_loss}")
```
