```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt


import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple
from ae_helper import get_data
from sklearn.model_selection import train_test_split
from torchvision import datasets, transforms
```

```python
# Load and prepare data
def get_data():
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
    test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=len(train_dataset), shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=len(test_dataset), shuffle=False)
    
    train_data = next(iter(train_loader))
    test_data = next(iter(test_loader))
    
    x_train, y_train = train_data
    x_test, y_test = test_data
    
    return (x_train.numpy(), y_train.numpy()), (x_test.numpy(), y_test.numpy())


def split_data(images, labels):
    """Split data into two parts with an equal number of each digit."""
    unique_labels = np.unique(labels)
    indices = np.arange(len(labels))
    part1_indices = []
    part2_indices = []

    for label in unique_labels:
        label_indices = indices[labels == label]
        np.random.shuffle(label_indices)
        mid = len(label_indices) // 2
        part1_indices.extend(label_indices[:mid])
        part2_indices.extend(label_indices[mid:])
    
    return images[part1_indices], labels[part1_indices], images[part2_indices], labels[part2_indices]

def create_pairs(images1, labels1, images2, labels2):
    """Create pairs where each digit in images1 is mapped to the next digit in images2."""
    x_pairs = []
    y_pairs = []
    for i in range(len(labels1)):
        current_image = images1[i]
        current_label = labels1[i]
        next_label = (current_label +1  ) % 10
        next_image_index = np.where(labels2 == next_label)[0][i % len(labels2[labels2 == next_label])]
        next_image = images2[next_image_index]
        x_pairs.append(current_image)
        y_pairs.append(next_image)
    return np.array(x_pairs), np.array(y_pairs)
# Load and prepare data
(X_train, y_train), (X_test, y_test) = get_data()

# Split the training data into two equal parts
X_train_part1, y_train_part1, X_train_part2, y_train_part2 = split_data(X_train, y_train)

# Create pairs
X_train, Y_train = create_pairs(X_train_part1, y_train_part1, X_train_part2, y_train_part2)
X_test_part1, y_test_part1, X_test_part2, y_test_part2 = split_data(X_test, y_test)
X_test, Y_test = create_pairs(X_test_part1, y_test_part1, X_test_part2, y_test_part2)

# Normalize and reshape the data correctly for PyTorch
X_train, X_test = [x.reshape(-1, 28, 28, 1) for x in (X_train, X_test)]
Y_train, Y_test = [y.reshape(-1, 28, 28, 1) for y in (Y_train, Y_test)]

# Further split X_train and Y_train into training and validation sets
x_train_pairs, x_test_pairs, y_train_pairs, y_test_pairs = train_test_split(X_train, Y_train, test_size=0.1, random_state=17)
```

```python
# Encoder
input_img = Input(shape=(28, 28, 1))
x = Flatten()(input_img)
x = Dense(128, activation='relu')(x)
encoded = Dense(64, activation='relu')(x)

# Decoder
x = Dense(128, activation='relu')(encoded)
x = Dense(28*28, activation='sigmoid')(x)
decoded = Reshape((28, 28, 1))(x)

# Autoencoder model
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

autoencoder.summary()
```

```python
autoencoder.fit(x_train_pairs, y_train_pairs,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test_pairs, y_test_pairs))
```

```python
# Generate next digit images
decoded_imgs = autoencoder.predict(x_test_pairs)

# Display the results
n = 10  # Number of images to display
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display original images
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test_pairs[i].reshape(28, 28), cmap='gray')
    plt.title(f"Original: {y_test[i]}")
    plt.axis('off')

    # Display generated images
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28), cmap='gray')
    plt.title(f"Next: {(y_test[i] + 1) % 10}")
    plt.axis('off')

plt.show()
```
